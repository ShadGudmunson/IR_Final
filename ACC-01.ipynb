## Process Analysis
Taking a look at processes first.

```powershell
python3 ../../volatility3/vol.py -f ../S039/Memory/ACC-01-dc8b5b3e.vmem windows.pstree.PsTree
```

Immediately, `svchost.exe` with PID 3040 is suspicious, as it ran fram `explorer.exe` and not from `services.exe` as most do. Let's take a look at its handles.

```powershell
python3 ../../volatility3/vol.py -f ../S039/Memory/ACC-01-dc8b5b3e.vmem windows.handles.Handles --pid 3040
```

Nothing jumps out as suspicious here. Let's take a look at DLLs.

```powershell
python3 ../../volatility3/vol.py -f ../S039/Memory/ACC-01-dc8b5b3e.vmem windows.dlllist.DllList --pid 3040
```

Nothing jumps out as immediately odd within those DLLs.

After taking a look at ACC-03, it seems `svchost.exe` was copied to the startup folder on this machine. Since taking a look at its DLLs and handles didn't seem to yield anything, it might help to dump the program and floss it.

```powershell
python3 ../../volatility3/vol.py -q -o ../svchost_dump -f ../S039/Memory/ACC-01-dc8b5b3e.vmem windows.pslist.PsList --pid 3040 --dump
```

Let's floss this program to get some insight into what's really going on.

```powershell
../../floss ../svchost_dump/pid.3040.0x11a0000.dmp
```
